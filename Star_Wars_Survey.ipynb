# Star Wars Survey

While waiting for Star Wars: The Force Awakens to come out, the team at FiveThirtyEight became interested in answering some questions about Star Wars fans. In particular, they wondered: does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?

The team needed to collect data addressing this question. To do this, they surveyed Star Wars fans using the online tool SurveyMonkey. They received 835 total responses, which you download from [their GitHub repository](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

The data has several columns, including:

- RespondentID - An anonymized ID for the respondent (person taking the survey)
- Gender - The respondent's gender
- Age - The respondent's age
- Household Income - The respondent's income
- Education - The respondent's education level
- Location (Census Region) - The respondent's location
- Have you seen any of the 6 films in the Star Wars franchise? - Has a Yes or No response
- Do you consider yourself to be a fan of the Star Wars film franchise? - Has a Yes or No response

```python
#explore the data
star_wars.head(10)
```

```python
star_wars.columns
```

```python
#Remove any data with RespondentID of null
star_wars = star_wars[star_wars["RespondentID"].notnull()].copy()
```

```python
star_wars.head(10)
```

```python
#Convert the columns with yes,no,NaN to boolean type True, False, NaN
yes_no = {"Yes":True, "No":False}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
#checking value count
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts()
```

```python
import numpy as np
#Convert whether respondent saw the series to true or false
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}
#Mapping to convert the values
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
#Change the column names for better readability
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
#convert values of rank to float type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })
```

```python
star_wars.head()
```

```python
ranking_mean=star_wars[star_wars.columns[9:15]].mean()
print(ranking_mean)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(range(6),ranking_mean)
```

## Rankings

So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each episode. As I suspected, it looks like the "original" movies are rated much more highly than the newer ones which means respondent favor the newer episodes.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

First episode had many respondents viewed and number decreased but from episode 4, it increased again and at episode 5, we have maximum viewers and also maximum rate as in previous bar chart. 

## View counts

```python
#split the data into two groups : males and females
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
#which episode the male and female prefer
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.title("Male")
plt.show()

plt.bar(range(6), females[females.columns[9:15]].mean())
plt.title("Female")
plt.show()
```

```python
#which episodes the male and female seen
plt.bar(range(6), males[males.columns[3:9]].mean())
plt.title("Male")
plt.show()

plt.bar(range(6), females[females.columns[3:9]].mean())
plt.title("Female")
plt.show()
```

## Conclusion

Interestingly, more males watches episodes 1-3, but males liked them far less than females did. And episode 5, seems to be the best episode in for both male and female in most-seen-wise and most-favored-wise.
